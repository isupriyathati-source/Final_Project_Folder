```python
# Step 0: Imports
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# You must define or import this function
# from your previous code: evaluate_quantiles
# from your earlier notebook or script
# Example placeholder:
def evaluate_quantiles(X_scaled, quantiles=[0.05,0.1,0.15,0.2,0.25]):
    # Placeholder: Replace with your actual implementation
    import pandas as pd
    results = []
    for q in quantiles:
        results.append({
            'quantile': q,
            'bandwidth': 1.0,       # dummy
            'n_clusters': 3,        # dummy
            'S': 0.5,               # dummy
            'CH': 300,              # dummy
            'DB': 0.2,              # dummy
            'composite': 0.5        # dummy
        })
    return pd.DataFrame(results)

# Step 1: Load and preprocess your dataset
df = pd.read_csv("customer_data.csv")

# Drop irrelevant columns
df.drop(columns=['CustomerID', 'Gender', 'Age', 'Spending Score (1-100)', 'Mail id'], inplace=True)

# Weights
WEIGHT_RFM = 0.8
WEIGHT_INCOME = 0.2

# Compute RFM Score
df['RFM Score'] = (0.2 * (100 - df['Recency (days)'])) + \
                  (0.4 * df['Frequency (visits)']) + \
                  (0.4 * df['Monetary ($)'])

# Weighted features
X = df[['RFM Score', 'Annual Income (k$)']].copy()
X['RFM Score'] *= WEIGHT_RFM
X['Annual Income (k$)'] *= WEIGHT_INCOME

# Step 2: Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Evaluate quantiles
results_df = evaluate_quantiles(X_scaled, quantiles=[0.05,0.1,0.15,0.2,0.25])
print(results_df[['quantile','bandwidth','n_clusters','S','CH','DB','composite']])
```
